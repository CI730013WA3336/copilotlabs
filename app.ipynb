```python
# create a dataframe of 100 sample employees with fieds name ,age and salary 
import pandas as pd
import random
import string
import numpy as np
# create a list of 100 random names
def random_name():
    return ''.join(random.choice(string.ascii_lowercase) for i in range(5))
names = [random_name() for i in range(100)]
# create a list of 100 random ages
ages = [random.randint(20,65) for i in range(100)]
# create a list of 100 random salaries
salaries = [random.randint(20000,80000) for i in range(100)]
# create a dictionary of the lists
data = {'name':names,'age':ages,'salary':salaries}
# create a dataframe from the dictionary
df = pd.DataFrame(data)
print(df.head())
print(df.describe())
print(df.info())
print(df['name'].head())



```

```python
# generate a bar chart of the salaries of the employees
import matplotlib.pyplot as plt
df['salary'].plot(kind='bar')
plt.show()
#generate a histogram of the salaries of the employees
df['salary'].plot(kind='hist')
plt.show()

```
